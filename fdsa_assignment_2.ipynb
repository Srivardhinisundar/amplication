```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
df=pd.read_csv("/content/Email.csv")
```

```python
df.head()
```

```python
df[['infrastructure','military']].describe()
```

```python

#value=146 refers to the null hypothesis value for the variable
ztest ,pvalue = stests.ztest(df['infrastructure'], x2=None, value=146)
print(float(pvalue))
print(ztest)
```

```python
if pvalue<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
     
```
